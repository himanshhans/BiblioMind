# downloading data from kaggle onto our device which also gives us the exact path to the data 

# data is downloaded as a csv file

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

# read_csv method is used to read the data as a pandas dataframe

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# now we are going to create a heatmap to check missing values
---
## A heatmap in Seaborn is a data visualization tool used to represent data in a matrix-like format, where individual values are displayed using varying color intensities. It's particularly helpful for visualizing patterns, correlations, or distributions in datasets.



```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.show()
```

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method= "spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman Correlation"})
heatmap.set_title("Correaltion Heatmap")

plt.show()
```

```python
book_missing = books[~(books["description"].isna()) &
      ~(books["num_pages"].isna()) &
      ~(books["average_rating"].isna()) &
      ~(books["published_year"].isna())
      ]
```

```python
book_missing
```

```python
df1 = book_missing["categories"].value_counts().reset_index().sort_values("count", ascending=False)
```

```python
df1
```

```python
bar_chart = df1.plot.bar(x="categories", y="count", color= "red", width=0.7, figsize=(105, 25))

plt.yscale("log")
plt.xlabel("Categories")
plt.ylabel("Count")
plt.title("Category Count Bar Chart")

plt.show()
```

### now final step in cleaning our data is that we need to refine the descriptions such that they are meaningful and have enough information so that we can recommend that book

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

### Now we will create a new data set with description length greater than 25 words

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words
```

### Now we will combine title and subtile fields as subtitle own its own is pretty much useless
---

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words["title"],
             book_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
book_missing_25_words
```

### Now we will create a new description column which tags each description with a unique identifier(isbn13) as it is important for building vector database
---

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25_words
```

### now we will drop some of the columns that we don't need and save this cleaned data into a csv file
---

```python
(
    book_missing_25_words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index=False)
)
```
